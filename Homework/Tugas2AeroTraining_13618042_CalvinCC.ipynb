## Kalkulasi Kecepatan <i>crash</i> kendaraan apabila diketahui <i> residual crush</i> kendaraan

### Penjelasan awal code
code ini merupakan **OOP-*based calculator*** sederhana untuk **menghitung kecepatan *crash* kendaraan.**<br>
code ini membentuk *class* "CrashVelocityCalculator" yang mana mengandung *property* dan *method* sebagai berikut:
<ol>
    <li> <i>Property</i> yang ingin didefinisikan untuk permasalah adalah:
        <ul> 
            <li> L<sub>0</sub> : panjang awal kendaraan sebelum <i>crash</i> [in] </li>
            <li> L<sub>1</sub> : panjang akhir kendaraan setelah <i>crash</i> [in] </li>
            <li> A  : konstanta kekakuan A [lbf/in] </li>
            <li> B  : konstanta kekakuan B [lbf/in<sup>2</sup>] </li>
            <li> G  : konstanta kekakuan G [lbf] </li>
            <li> W  : lebar kendaraan [in] </li>
            <li> w  : berat kendaraan [lbf] </li>
        </ul><br>
    <li> <i>method</i> yang ingin didefinisikan untuk permasalah adalah:
        <ul> 
            <li> perhitungan <i>residual crush</i></li>
            <li> perhitungan <i>crush energy</i></li>
            <li> perhitungan kecepatan <i>crash</i> kendaraan</li>
        </ul>

```python
class CrashVelocityCalculator():
    def __init__(self, L0, L1, A, B, G, W, w):
        self.L0 = L0
        self.L1 = L1
        self.A = A
        self.B = B
        self.G = G
        self.W = W
        self.w = w
    
    def residual_crush(self):
        return self.L0 - self.L1
    
    def crush_energy(self):
        C = self.L0 - self.L1
        return self.W*(self.G + self.A*C + (self.B/2)*C**2)
    
    def crash_velocity(self):
        b0 = 1.12*self.A*(self.W/(self.B*self.w))**(1/2)
        b1 = 1.12*(self.B*self.W/self.w)**(1/2)
        C = self.L0 - self.L1
        return b0 + b1*C
```

    <li>Untuk ketiga konstanta kekakuan kendaraan <b>A</b>, <b>B</b>, dan <b>G</b> dapat ditentukan berdasarkan tipe <i>crash</i> dan jenis kendaraan pada tabel berikut:</li>
</ul>

</ul>

### Perhitungan dan menampilkan hasil
pada <i>cell</i> berikut, maka <i>user</i> akan diminta input nilai <i>property</i> L0, L1, A, B, G, W, dan w yang telah diperoleh. Lalu, nilai <i>property</i> tersebut akan dimasukkan pada <i>class</i> yang terbentuk sebelumnya, dilakukan perhitungan menggunakan <i>method</i> yang telah didefinisikan, dan menampilkan hasil tersebut.

```python
L0 = float(input("Masukkan nilai L0 [in]: "))
L1 = float(input("Masukkan nilai L1 [in]: "))
A = float(input("Masukkan nilai A [lbf/in]: "))
B = float(input("Masukkan nilai B [lbf/in^2]: "))
G = float(input("Masukkan nilai G [in]: "))
W = float(input("Masukkan nilai W [in]: "))
w = float(input("Masukkan nilai w [lbf]: "))
data_kendaraan = CrashVelocityCalculator(L0, L1, A, B, G, W, w)
print("maka,")
print("residual crush kendaraan adalah {} in".format(data_kendaraan.residual_crush()))
print("crush energy kendaraan adalah {} lbf/in".format(data_kendaraan.crush_energy()))
print("kecepatan crash kendaraan adalah {} mph".format(data_kendaraan.crash_velocity()))
```
